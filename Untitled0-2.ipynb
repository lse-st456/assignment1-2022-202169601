```python
from tensorflow.keras import Model
from tensorflow.keras import Input
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras import layers
import tensorflow as tf
import keras
import matplotlib.pyplot as plt
import pandas
import numpy as np
import statistics

```

```python
#p1,1
def my_model_make(lat_dim,learning_rate=1e-4):
  my_model = keras.Sequential()
  my_model.add(layers.Dense(100,activation='relu'))
  #this is a hidden layer with 100 neurons and activation function relu.
  my_model.add(layers.Dense(lat_dim))
  #this is a hidden layer with lat_dim neurons.
  #the two layers above are function phi
  my_model.add(layers.Dense(100,activation='relu'))
  #this is a hidden layer with 100 neurons and activation function relu.
  my_model.add(layers.Dense(1))
  #this is an output layer
  #the two layers above are function rho.
  my_model.compile(
    optimizer=tf.optimizers.SGD(learning_rate=learning_rate),loss='mean_squared_error',metrics='accuracy')
  return my_model
```

```python
def my_model_fit(model,x_train,y_train,epochs):
  fitted_model=model.fit(
    x_train,
    y_train,
    validation_split=0.1,
    epochs=epochs,
    batch_size=128,
    verbose=0,)
  return fitted_model
```

```python
#input my data here
from matplotlib import test
test_data=pandas.read_csv('/content/test-1.csv',header=None,index_col=False).values
train_data=pandas.read_csv('/content/train-1.csv',header=None,index_col=False).values

test_x=test_data[:,0:10]
test_y=test_data[:,10]
train_x=train_data[:,0:10]
train_y=train_data[:,10]



```

```python
#p1.2
model=my_model_make(lat_dim=5)
trained_model=my_model_fit(model=model,x_train=train_x,y_train=train_y,epochs=10)
```

```python
plt.plot(trained_model.history['loss'])
plt.plot(trained_model.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['test', 'validation'], loc='upper right')
plt.show()
print("the test MSE value is ",model.evaluate(test_x,test_y,verbose=0)[0])
```

```python
model=my_model_make(lat_dim=100)
trained_model=my_model_fit(model=model,x_train=train_x,y_train=train_y,epochs=10)
plt.plot(trained_model.history['loss'])
plt.plot(trained_model.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
print("the test MSE value is ",model.evaluate(test_x,test_y,verbose=0)[0])
```

The two models have different "lat_dim", both plots shows a stricktly decreasing trend, the validation loss value and the MSE value of the model with "lat_dim=100" is smaller. Hence, the second model is better.

```python
#p1.4
def my_RELUmodel_make(lat_dim,learning_rate=1e-4):
  my_model = keras.Sequential()
  my_model.add(layers.Dense(100,activation='relu'))
  my_model.add(layers.Dense(lat_dim,activation='relu'))
  my_model.add(layers.Dense(100,activation='relu'))
  my_model.add(layers.Dense(1,activation='relu'))
  my_model.compile(
    optimizer=tf.optimizers.SGD(learning_rate=learning_rate),loss='mean_squared_error',metrics='accuracy')
  return my_model
relumodel=my_RELUmodel_make(lat_dim=100,learning_rate=1e-4)
trained_relumodel=my_model_fit(model=relumodel,x_train=train_x,y_train=train_y,epochs=50)
plt.plot(trained_relumodel.history['loss'])
plt.plot(trained_relumodel.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()

```

```python
def my_sigmoidmodel_make(lat_dim,learning_rate=1e-4):
  my_model = keras.Sequential()
  my_model.add(layers.Dense(100,activation='relu'))
  my_model.add(layers.Dense(lat_dim,activation='sigmoid'))
  my_model.add(layers.Dense(100,activation='relu'))
  my_model.add(layers.Dense(1,activation='sigmoid'))
  my_model.compile(
    optimizer=tf.optimizers.SGD(learning_rate=learning_rate),loss='mean_squared_error',metrics='accuracy')
  return my_model
sigmoidmodel=my_sigmoidmodel_make(lat_dim=100,learning_rate=1e-4)
trained_sigmoidmodel=my_model_fit(model=sigmoidmodel,x_train=train_x,y_train=train_y,epochs=50)
plt.plot(trained_sigmoidmodel.history['loss'])
plt.plot(trained_sigmoidmodel.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

the relu model decreases faster and ends at a lower validation loss arround 0.01, so the relu model here is better.

```python
def my_model2_make(L,w):
  my_model = keras.Sequential()
  while L>1:
    my_model.add(layers.Dense(w,activation='relu'))
    L-=1#use a while loop to create L-1 layers and the output dimension is m*w
  my_model.add(layers.Dense(1))
  #the output dimension is m*1
  my_model.compile(optimizer=tf.optimizers.Adam(learning_rate=1E-4,epsilon=1e-3),loss='mean_squared_error',metrics=['accuracy'])
  return my_model
```
